#### 2. Write some helper functions for stack

```python
import timeit
# Function to create an empty stack.
# It initializes size of stack as 0
def createStack():
    stack = []
    return stack
 
# Function to determine the size of the stack
def size(stack):
    size = len(stack)
    return size
 
# Function to test if the Stack is empty
# Stack is empty if the size is 0
def isEmpty(stack):
    if size(stack) == 0:
        return True
    else:
        return False
 
# Function to add an item to stack . 
# It increases size by 1 
def push(stack,item):
    stack.append(item)
 
#Function to remove an item from stack. 
# It decreases size by 1
def pop(stack):
    stack.pop()
```

#### 3. Write the reverse string function

```python
# A stack based function to reverse a string
def reverse(string):
    # Create my stack
    rev_stack = createStack()

    # Create a loop, iterate over each letter, add letter to stack
    for letter in string:
        rev_stack.append(letter)

    # Pull each letter out of the stack to add to a string
    rev_string = []
    while rev_stack:
        rev_string.append(rev_stack.pop())
    return ''.join(rev_string)
     
```

### Queue using Stacks
Given a stack data structure with push and pop operations implement a queue using instances of stack data structure and operations on them.
Hint: To implement a queue you have to define the $enQueue(q, x)$ and $deQueue(q)$ operations using stacks.

#### 1. Write an algorithm for this task that has costly $enQueue(q,x)$ operation

```python
# Function to create an empty stack.
# It initializes size of stack as 0
def createStack1():
    return []

# Function to determine the size of the stack
def size1(stack):
    return  len(stack)

# Function to test if the Stack is empty
# Stack is empty if the size is 0
def isEmpty1(stack):
    return size1(stack) == 0

# Function to add an item to stack .
# It increases size by 1
def push1(stack,item):
    stack.append(item)

#Function to remove an item from stack.
# It decreases size by 1
def pop1(stack):
    return stack.pop()

def enqueue_costly(q,x): #costly
    # Creates a temporary stack
    temp = createStack1()
    # Moves everything in stack to temp stack
    while not isEmpty1(q):
        push1(temp, pop1(q))
    #Add the item to the bottom of the q stack
    push1(q,x)
    #Move back to q
    while not isEmpty1(temp):
        push1(q, pop1(temp))
    return q

def dequeue_cheap(q): #Not costly
    if isEmpty1(q):
        return None
    return pop1(q)

#Test
#Create Empty Stack
s = createStack1()

#Build Queue
enqueue_costly(s,4)
enqueue_costly(s,3)
enqueue_costly(s,2)
enqueue_costly(s,1)

#Should return 4,3,2,1 because that's the order they were added in
while not isEmpty1(s):
    print(dequeue_cheap(s))
```

#### 2. Write an algorithm for this task that has costly $deQueue(q)$ operation

```python
#Using same helper functions as the above code i.e isEmpty1

def enqueue_cheap(q,x): #cheap
    push1(q,x)

def dequeue_costly(q): #costly
    if isEmpty1(q):
        return None
    # Creates a temporary stack
    temp = createStack1()
    # Moves everything in stack to temp stack
    while not isEmpty1(q):
        push1(temp, pop1(q))
    #Dequeue
    front = pop1(temp)
    #Put everything back in q
    while not isEmpty1(temp):
        push1(q, pop1(temp))

    return front

#Test
#Create Empty Stack
s = createStack1()

#Build Queue
enqueue_cheap(s,4)
enqueue_cheap(s,3)
enqueue_cheap(s,2)
enqueue_cheap(s,1)

#Should return 4,3,2,1 because that's the order they were added in
while not isEmpty1(s):
    print(dequeue_costly(s))
```

#### 4. A word on time complexity of those operations in each one of the above cases

## Cheap Enqueue Costly Dequeue
In this example the cheap enqueue is O(1) because it's just a regular .append function which is constant-time. The costly dequeue is O(n) because it moves all of the contents of one stack into another, which would increase time complexity dependent on the amount of contents.

## Cheap Dequeue Costly Enqueue
In this example the cheap dequeue is 0(1) because the enqueue function changes the order so that the oldest item is the one on the top. So all it needs to do is .pop() which is O(1). The enqueue is O(n) because it is moving all the items into another stack to add in the new item in the right place.
