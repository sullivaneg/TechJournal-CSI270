### Here is how we define a node class

```python
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self,newdata):
        self.data = newdata

    def setNext(self,newnext):
        self.next = newnext
```

# TODO: Write some code that adds a new element to the list. Try your best to make it O(1)

```python
# Code and have fun here, don't overthink, it's easier than you think
class Node:
    def __init__(self,initdata):
        self.data = initdata
        self.next = None

class UnorderedList:
    def __init__(self):
        self.head = None

    def add(self, data):
        # Create my node
        new_node = Node(data)
        # Move current head to next
        if self.head is None:
            # If list is empty make the current node the new head
            self.head = new_node
        else:
            # If list isn't empty move the current head to the next spot
            new_node.next = self.head
            # and make the new node the head
            # This makes it O(1) because it doesn't have to transverse the list
            self.head = new_node

    # Playing around with printing
    def print_list(self):
        start = self.head
        while start:
            print(start.data, end=" ")
            start = start.next
```
